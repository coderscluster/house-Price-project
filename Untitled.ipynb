```python
import pandas as pd
df= pd.read_excel("C:\\Users\\RAM\\Downloads\\innercity.xlsx")
df.head()
```

```python
df['dayhours']=df['dayhours'].map(lambda x: str(x)[:-7])
```

```python
#preprocessing
df.isnull().sum()
```

```python
df.info()
```

```python
df['living_measure']=df['living_measure'].fillna(df['living_measure'].mean())
df['lot_measure']=df['lot_measure'].fillna(df['lot_measure'].mean())
df['living_measure15']=df['living_measure15'].fillna(df['living_measure15'].mean())
df['lot_measure15']=df['lot_measure15'].fillna(df['lot_measure15'].mean())
df['ceil_measure']=df['ceil_measure'].fillna(df['ceil_measure'].mean())
df['basement']=df['basement'].fillna(df['basement'].mean())
df['yr_built']=df['yr_built'].fillna(df['yr_built'].mode()[0])
df['quality']=df['quality'].fillna(df['quality'].mode()[0])
df['ceil']=df['ceil'].fillna(df['ceil'].mode()[0])
df['room_bed']=df['room_bed'].fillna(df['room_bed'].mode()[0])
df['room_bath']=df['room_bath'].fillna(df['room_bath'].mode()[0])
df['condition']=df['condition'].fillna(df['condition'].mode()[0])
```
