```python
import numpy as np
import pandas as pd
df= pd.read_excel("C:\\Users\\RAM\\Downloads\\innercity.xlsx")
df.head()
```

```python
df=df.dropna()
assert df.isna().sum().sum()==0
```

```python
df.isna().any()
```

```python
df.shape
```

```python
df=pd.get_dummies(df,columns=['zipcode'])
```

```python
df.shape
```

```python
np.random.seed(1)
train=np.random.rand(len(df))<0.7
price='price'
X_train=df[train].drop(price, axis=1)
X_test=df[~train].drop(price, axis=1)
y_train=df[train][price]
y_test=df[~train][price]

display(X_train.head())
X_train.shape
```

```python
def mse(y_hat,y):
    return np.sum(np.square(y_hat-y))/y.size

f='{}~{}'.format(price,'+'.join(X_train.columns))
import statsmodels.formula.api as smf

model=smf.ols(formula=f, data=pd.concat([X_train,y_train],axis=1)).fit()
y_hat=model.predict(X_test)

print('RMSE {}'.format(np.sqrt(mse(y_hat,y_test))))
print('RSquared',model.rsquared())
```
