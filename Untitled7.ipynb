```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df= pd.read_excel("C:/Users/RAM/Downloads/innercity.xlsx")
df.info()
```

```python
df.drop(['cid','dayhours'],axis=1,inplace=True)

for columns in df.columns:
    df[columns]=df[columns].replace('$',np.nan)
    
df.isnull().sum()
```

```python
#handling with null values
columns=['price', 'room_bed', 'room_bath', 'living_measure',
       'lot_measure', 'ceil', 'coast', 'condition', 'quality',
       'ceil_measure', 'basement', 'yr_built', 'yr_renovated',
       'lat', 'long', 'living_measure15', 'lot_measure15', 'furnished','sight','total_area']
for i in columns:
    df[i]=df[i].fillna(df[i].median())
    
#double checking NaN
df.isnull().any()
```

```python
df.drop(['living_measure','lot_measure'], axis=1,inplace=True)

x=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=18)
```

```python
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor()
cv=KFold(n_splits=10,random_state=0,shuffle=False)
y_pred=cross_val_predict(model, X_train, y_train, cv=cv)

print('RSquared: '+str(round(r2_score(y_train,y_pred),2)))
```

```python
model.fit(X_train,y_train)
```

```python
y_pred2=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
